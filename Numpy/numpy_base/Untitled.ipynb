# `2)ndarray的创建`

### `2.1)np.ones(shape, dtype=None, order='C')`
`Return a new array of given shape and type, filled with ones.`

```python
ones_ndarray = np.ones(shape=(10,8), dtype=int)
print(type(ones_ndarray))
plt.imshow(ones_ndarray)
```

### `2.2)np.zeros(shape, dtype=float, order='C')`
`Return a new array of given shape and type, filled with zeros.`

```python
zeros_ndarray = np.zeros(shape=(10,8), dtype=int)
plt.imshow(zeros_ndarray)
plt.show()
```

### `2.3)np.full(shape, fill_value, dtype=None, order='C')`
`Return a new array of given shape and type, filled with 'fill_value'.`

```python
full_ndarray = np.full(shape=(4,4), fill_value=6, dtype=int)
plt.imshow(full_ndarray)
plt.show()
```

### `2.4)np.eye(N, M=None, k=0, dtype=<class 'float'>, order='C')`
`Return a 2-D array with ones on the diagonal and zeros elsewhere.`

```python
eye_ndarray = np.eye(5)
plt.imshow(eye_ndarray)
plt.show()
```

### `2.5)np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)`
`Return evenly spaced numbers over a specified interval.`

```python
linspace_ndarray = np.linspace(0, 100, 20)
print(linspace_ndarray)
```

### `2.6)arange([start,] stop[, step,], dtype=None)`
`Return evenly spaced values within a given interval.`

```python
arange_ndarray = np.arange(0, 100, 20)
print(arange_ndarray)
```

### `2.7)randint(low, high=None, size=None, dtype='l')`
`Return random integers from 'low' (inclusive) to 'high' (exclusive).`

```python
randint_ndarray = np.random.randint(100, 200, size=5)
print(randint_ndarray)
```

### `2.8)randn(d0, d1, ..., dn)`
`Return a sample (or samples) from the "standard normal" distribution.`

```python
randn_ndarray = np.random.randn(100)
print(randn_ndarray)
```

### `2.9)normal(loc=0.0, scale=1.0, size=None)`
`Draw random samples from a normal (Gaussian) distribution`

```python
normal_ndarray = np.random.normal(175, 2, 10)
print(normal_ndarray)
```

### `2.10)random_sample(size=None) `
`Return random floats in the half-open interval [0.0, 1.0).`

```python
random_sample_ndarray = np.random.random_sample(size = (200, 300, 3))
plt.imshow(random_sample_ndarray)
plt.show
```

# `3)ndarray的基本操作`

### `3.1)索引`

```python
n1 = np.random.randint(0, 100, size=(3, 4))
print(n1)
print(n1[1, 1])
```

```python
n1 = np.random.randint(0, 100, size=(3, 4, 5))
print(n1)
print(n1[0, 1, 1])
```

### `3.2)切片`

```python
test_ndarray = np.random.randint(0, 100, size=(3, 4, 5))
print(test_ndarray)
```

```python
print(test_ndarray[2:3])
```

```python
print(test_ndarray[2:3,0:1])
```

```python
print(test_ndarray[2:3, 0:1, -3:])
```

```python
print(test_ndarray[::-1, ::-1, ::-1])
```

### `3.3)变形`
`np.reshape(a, newshape, order='C')`

```python
# resahpe 如果接受负数，则直接转换为1维数组
cat3 = cat.reshape(-1)
print(cat3)
```

### `3.4)级联`
### `3.4.1)concatenate((a1, a2, ...), axis=0, out=None)`

`Join a sequence of arrays along an existing axis.`

```python
test_ndarray = np.random.randint(0, 10, size=(3, 4, 5))
print(test_ndarray)
```

```python
result_ndarray = np.concatenate((test_ndarray, test_ndarray), axis=0)
print(result_ndarray)
```

```python
result_ndarray = np.concatenate((test_ndarray, test_ndarray), axis=1)
print(result_ndarray)
```

```python
result_ndarray = np.concatenate((test_ndarray, test_ndarray), axis=2)
print(result_ndarray)
```

```python
result_cat = np.concatenate((cat, cat), axis=1)
plt.imshow(result_cat)
plt.show
```

### `3.4.2)np.vstack(tup)`
`Stack arrays in sequence vertically (row wise).`

```python
print(test_ndarray)
print("*" * 30)
result_ndarray = np.vstack(test_ndarray)
print(result_ndarray)
```

### `3.4.2) np.hstack(tup)`
`Stack arrays in sequence horizontally (column wise).`

```python
print(test_ndarray)
print("*" * 30)
result_ndarray = np.hstack(test_ndarray)
print(result_ndarray)
print("*" * 30)
result_ndarray = np.hstack(result_ndarray)
print(result_ndarray)
```

### `3.5)副本`
#### `对赋值后对象的操作也会对原来的对象生效`

```python
n1 = np.array([1, 2, 3, 4])
n2 = n1
n2[0] = 10
print(n1, n2)
```

#### `可用copy函数创建副本`
`np.copy(a, order='K')`

`Return an array copy of the given object.`


```python
n1 = np.array([1, 2, 3, 4])
n2 = n1.copy()
n2[0] = 10
print(n1, n2)
```

# `5)ndarray的矩阵操作`

#### `5.1)矩阵积`
`dot(a, b, out=None)`

`Dot product of two arrays. Specifically,`

```python
n1 = np.random.randint(0, 10, size=(2, 3))
n2 = np.random.randint(0, 10, size=(3, 2))
print(n1)
print(n2)
print(np.dot(n1, n2))
```

# `6)ndarray的排序操作`

```python
test_ndarray = np.random.randint(0, 100, size=(3, 4, 5))
print(test_ndarray)
```

### `6.1)np.sort(a, axis=-1, kind='quicksort', order=None)`

`Return a sorted copy of an array.`

```python
print(np.sort(test_ndarray, axis=0))
```

### `6.2)a.sort(axis=-1, kind='quicksort', order=None)`

`Sort an array, in-place.`


```python
test_ndarray.sort(axis=1)
print(test_ndarray)
```

### `6.3)np.partition(a, kth, axis=-1, kind='introselect', order=None)`
`Return a partitioned copy of an array.`

```python
test_ndarray = np.random.randint(0, 100, size=(10))
print(np.partition(test_ndarray, kth=5)[:5])
print(np.partition(test_ndarray, kth=-5)[-5:])
```
